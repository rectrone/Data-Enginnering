```python

import requests
```

```python
url="https://a.4cdn.org/boards.json"
```

```python
res=requests.get(url)
```

```python
data=res.json()
```

```python
data['boards'][0]
```

```python
boards_data=[]
for boardData in data['boards']:
    board_title=boardData['title']
    board_page=boardData['pages']
    board_filesize=boardData['max_filesize']
    board_image_limit=boardData['image_limit']
    board_elem={
        "board_title":board_title,
        "board_page":board_page,
        "board_elem":board_page,
        "board_filesize":board_filesize,
        "board_image_limit":board_image_limit
    }
    boards_data.append(board_elem)
    
```
